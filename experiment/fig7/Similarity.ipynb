```python
from datasets import DataProvider
dataset_params = {
    'name': 'B100',
    'test_only': True,
    'patch_size': 192,
    'ext': 'sep',
    'scale': 4,
    "batch_size": 1,
}
provider = DataProvider(dataset_params)
```

```python
def get_sim(model1, model2, data_loader):
    import torch
    from tqdm import tqdm
    from frameworks.nnmetric.feature_similarity_measurement import cka_loss
    meter = cka_loss()
    model1 = model1.cuda()
    model2 = model2.cuda()
    with torch.no_grad():
        all_ret = []
        for batch in tqdm(data_loader):
            x, y, _ = batch
            x = x.cuda()
            feat1, _ = model1(x, with_feature=True)
            feat2, _ = model2(x, with_feature=True)
            ret = []
            for f1, f2 in zip(feat1[:-1], feat2[:-1]):
                c1 = f1.size(1)
                c2 = f2.size(1)
                f1 = f1.permute((0,2,3,1)).reshape((-1,c1))
                f2 = f2.permute((0,2,3,1)).reshape((-1,c2))
                ret.append(1-meter(f1,f2).item())
            all_ret.append(ret)
        return torch.tensor(all_ret).mean(dim=0)
```

```python
sim1 = get_sim(teacher, baseline, provider.test_dl)
sim2 = get_sim(teacher, plain, provider.test_dl)
sim3 = get_sim(teacher, init, provider.test_dl)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plot_x = [i+1 for i in range(len(sim1))]
font = {'family': 'Times New Roman', 'weight': 'normal', 'size': 18}
matplotlib.rc("font", **font)
plt.plot(plot_x, sim1, label='Baseline')
plt.plot(plot_x, sim2, label='Init.+Distill')
plt.plot(plot_x, sim3, label='Init.')
plt.legend(loc='best')
plt.savefig('sim'+".svg", dpi=1000, bbox_inches='tight')
plt.show()
```
