## Main

```python
def RGB_to_gray(preds):
    gray_coeffs = [65.738, 129.057, 25.064]
    convert = preds.new_tensor(gray_coeffs).view(1, 3, 1, 1) / 256
    return preds.mul(convert).sum(dim=1)
```

```python
def view(data, do_sum=False, save_to=None):
    import matplotlib.pyplot as plt
    %matplotlib inline
    if do_sum:
        plt.imshow(data.squeeze().permute(1,2,0).sum(dim=2).numpy()/255)
    else:
        plt.imshow(data.squeeze().permute(1,2,0).numpy()/255)
    plt.show()
    if save_to is not None:
        from PIL import Image
        ndarr = data.squeeze().clamp(0, 255).permute(1,2,0).byte().numpy()
        im = Image.fromarray(ndarr)
        im.save(save_to)

from datasets import DataProvider
dataset_params = {
    'name': 'Urban100',
    'test_only': True,
    'patch_size': 192,
    'ext': 'sep',
    'scale': 4,
    "batch_size": 1,
}
provider = DataProvider(dataset_params)
```

```python
name = ('plain_x', 'plain_m', 'imdn', 'edsr', 'bib', 'carn')
methods = (plain_x, plain_m, imdn, edsr, bib, carn)
coe = (-1, -1, 1, 1, 0, 1)
# coe = (1, 1, -1, -1, 0, -1)

best_lr = None
best_hr = None
best_score = -1e9
best_mp = None
with torch.no_grad():
    for idx, batch in enumerate(provider.test_dl):
        x, y, _ = batch
        x = x.cuda()
        gray_y = RGB_to_gray(y).cuda()
        mp = torch.zeros_like(gray_y)
        for m, c in zip(methods, coe):
            m = m.cuda()
            p = RGB_to_gray(m(x))
            mp += (gray_y-p).abs() * c
        mp = mp.cpu()
        score = mp.max()
        if score > best_score:
            best_score = score
            best_hr = y.cpu()
            best_lr = x.cpu()
            best_mp = mp[0]
            print('update with img_id = ', idx, 'score = ', score, 'filename = ', _)
```

```python
window_size = (40, 80)
# window_size = (80, 200)
width = best_mp.size(1)
pos = best_mp.view((-1,)).argmax(0).item()
pos_r, pos_c = pos//width, pos%width
print(pos_r, pos_c)
selects = ( 0,
            slice(0, None),
            slice(max(0, pos_r-window_size[0]), min(best_mp.size(0), pos_r+window_size[0])),
            slice(max(0, pos_c-window_size[1]), min(best_mp.size(1), pos_c+window_size[1])),)
view(best_hr)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

plt.rcParams['figure.figsize'] = [16, 8]

ndarr = best_lr.squeeze().clamp(0, 255).permute(1,2,0).byte().numpy()
im = Image.fromarray(ndarr)

# Create figure and axes
fig, ax = plt.subplots()
# Display the image
ax.imshow(im)
# Create a Rectangle patch
rect = patches.Rectangle(((pos_c - window_size[1])//4, (pos_r - window_size[0])//4), window_size[1]//2, window_size[0]//2, 
                         linewidth=3, edgecolor='r', facecolor='none')
# Add the patch to the Axes
ax.add_patch(rect)
plt.axis('off')
plt.savefig('lr.png', bbox_inches='tight', dpi=1000)
plt.show()
```

```python
print('hr')
view(best_hr[selects], save_to='ex_hr.png')
with torch.no_grad():
    for m, n in zip(methods, name):
        m = m.cuda()
        out = m(best_lr.cuda())
        print(n)
        view(out.cpu()[selects], save_to=f'ex_{n}.png')
```
